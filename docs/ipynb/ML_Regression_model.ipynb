+ def **specs( *`Año`: str* )**:
    Se ingresa un año y devuelve una lista con los 5 specs que más se repiten en el mismo en el orden correspondiente. 

```python
from ast import literal_eval

def get_specs_dict(year):
  df = df_games.copy()
  specs_dict = {}
  
  def read_specs(specs):
    if isinstance(specs, str):

      for spec in list(literal_eval(specs)):
        if spec in specs_dict:
          specs_dict[spec] += 1
        else:
          specs_dict[spec] = 1
          
  df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
  
  df = df[df['release_date'] == int(year)]
  df['specs'].map(lambda x: read_specs(x))
  return {k: v for k, v in sorted(specs_dict.items(), key=lambda item: item[1], reverse =True)}

response = get_specs_dict(2018)
response
```

def **sentiment( *`Año`: str* )**:
    Según el año de lanzamiento, se devuelve una lista con la cantidad de registros que se encuentren categorizados con un análisis de sentimiento. 

```python
def get_sentimental_dict(year):
    df = df_games.copy()
    sentimental_dict = {}
    
    def read_sentimental(sentimental):
        if sentimental in sentimental_dict:
            sentimental_dict[sentimental] += 1
        else:
            sentimental_dict[sentimental] = 1
            
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
    df = df[df['release_date'] == int(year)]
    df['sentiment'].map(lambda x: read_sentimental(x))
    return sentimental_dict

response = get_sentimental_dict(2016)
response
```

+ def **earlyacces( *`Año`: str* )**:
    Cantidad de juegos lanzados en un año con early access.

```python
def get_early_access_amount(year):
    df = df_games[df_games['early_access'] == True].copy()
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
    df = df[df['release_date'] == int(year)]
    return df
    
prueba = get_early_access_amount(2018)
prueba.shape[0]
```

+ def **metascore( *`Año`: str* )**:
    Top 5 juegos según año con mayor metascore.

```python
def get_metascore(year: int, limit=None):
    df = df_games.copy()
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
    df = df[df['release_date'] == int(year)].sort_values(by=['metascore'], ascending=False)
    return df
    
prueba = get_metascore(2017)
prueba
```
